## Setup SST  Service

```python
apikey = 'your api key'
url = 'your url'
```

## Setup service

```python
authenticator = IAMAuthenticator(apikey)
stt = SpeechToTextV1(authenticator=authenticator)
stt.set_service_url(url)
```

```python
# Full path to the audio file
audio_file_path = '/content/Rising from Within [XoWRjVKGXK8].webm'  # Ensure this is the correct path
```

## Open Audio Source and Convert

```python
# Transcribe the audio file
try:
    with open(audio_file_path, 'rb') as f:
        res = stt.recognize(
            audio=f,
            content_type='audio/webm',  # Correct content type for .webm file
            model='en-AU_NarrowbandModel',  # Use the Australian English Narrowband model
            timestamps=True,  # Optional parameters
            word_confidence=True,  # Optional parameters
            max_alternatives=3  # Optional parameters
        ).get_result()

    # Print the results
    print(res)

except FileNotFoundError:
    print(f"The file at path {audio_file_path} was not found. Please check the path and try again.")
```
